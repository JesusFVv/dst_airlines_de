```python
import requests

import pandas as pd
```

```python
# Using direct communication between containers inside the same Network
headers = {'Accept-Profile': 'l2'}  # Define the schema of the table (l2 or l3)
r = requests.get('http://postgrest_api:3000/refdata_cities', headers=headers)
print(r.status_code)
print(r.json())
```

```python
# Using NGINX reverse proxy inside the same Network
headers = {'Accept-Profile': 'l2'}  # Define the schema of the table (l2 or l3)
r = requests.get('http://nginx:80/postgrest_api/refdata_cities', headers=headers)
print(r.status_code)
print(r.json())
```

```python
# Using NGINX reverse proxy from outside the Network, using the docker Gateway IP (for the default bridge network) (172.17.0.1)
# Whitout header specified it takes the l2 schema by default
r = requests.get('http://172.17.0.1:8000/postgrest_api/refdata_cities')
print(r.status_code)
print(r.json())
```

```python
# Using NGINX reverse proxy from outside the Network, using the docker Gateway IP for the project Network dst_network (172.19.0.1)
r = requests.get('http://172.19.0.1:8000/postgrest_api/refdata_cities')
print(r.status_code)
print(type(r.json()))
print(type(r.json()[0]))
print(r.json())
```
